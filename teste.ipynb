```python
import pandas as pd
import numpy as np
import re
import os # Necess√°rio para o FileNotFoundError

# --- 1. Carregamento dos DataFrames ---

# DataFrame Original (Onde a fundamentacao est√° no texto)
coluna_original='fundamentacao'
# DataFrame Secund√°rio: fundamentacao (Tabela de Lookup com Artigo, Numero_n, Alinea e ID)
try:
    # Assumimos que o ficheiro 'fundamentacao.csv' J√Å FOI PR√â-PROCESSADO
    # e contem as colunas 'Artigo', 'Numero_n', 'Alinea' e 'ID'.
    df_secundario_fund = pd.read_csv('tabelas2/fundamentacao.csv') 
    id_coluna_secundaria = 'ID' # Coluna com o ID na tabela secund√°ria
    colunas_chave_fund = ['Artigo', 'Numero_n', 'Alinea']
    print(f"‚úÖ DataFrame 'fundamentacao.csv' carregado para lookup. Colunas: {df_secundario_fund.columns.tolist()}")
except FileNotFoundError:
    print("‚ùå ERRO: O arquivo 'fundamentacao.csv' n√£o foi encontrado.")
    exit()

print("-" * 50)


# =========================================================================
# === TRANSFORMA√á√ÉO: Mapeamento de fundamentacao (Usando Extra√ß√£o Separada) ===
# =========================================================================

# --- Configura√ß√µes da sua extra√ß√£o ---
REGEX_ARTIGO = r'Artigo\s*(\d+\.¬∫|\d+)'
REGEX_NUMERO = r'n\.¬∫\s*(\d+)'
REGEX_ALINEA = r'al√≠nea\s*([a-z])'
novas_colunas_temp = ['Artigo', 'Numero_n', 'Alinea'] # Chaves de merge

# PR√â-REQUISITO: Criar um DataFrame de exemplo para o c√≥digo poder ser executado
# (Se o seu df_original estiver carregado a partir de ficheiro, pode remover esta parte)

# FIM DO PR√â-REQUISITO


# Pr√©-processamento do df_secundario_fund (Garantir que as chaves sejam string e vazias)
# IMPORTANTE: Se as colunas de Artigo/Numero_n no df_secundario_fund forem Int,
# ajuste esta linha para usar .astype(pd.Int64Dtype()).fillna(pd.NA)
df_secundario_fund[colunas_chave_fund] = df_secundario_fund[colunas_chave_fund].fillna('').astype(str).apply(lambda x: x.str.strip())


print("üëâ 1/4. Extraindo Artigo, N√∫mero e Al√≠nea de 'fundamentacao' (Sua l√≥gica)...")

# 1. Aplicar a Sua L√≥gica de Extra√ß√£o no df_original
df_original[coluna_original] = df_original[coluna_original].astype(str).str.strip()

df_original['Artigo'] = df_original[coluna_original].str.extract(REGEX_ARTIGO, flags=re.IGNORECASE)
df_original['Numero_n'] = df_original[coluna_original].str.extract(REGEX_NUMERO, flags=re.IGNORECASE)
df_original['Alinea'] = df_original[coluna_original].str.extract(REGEX_ALINEA, flags=re.IGNORECASE)

# 2. Limpeza P√≥s-Extra√ß√£o (C√ìDIGO MODIFICADO ABAIXO)
# Limpar Artigo para remover o '.¬∫'
df_original['Artigo'] = df_original['Artigo'].str.replace('\.¬∫', '', regex=True).str.strip()
df_original['Numero_n'] = df_original['Numero_n'].str.strip()
df_original['Alinea'] = df_original['Alinea'].str.strip()

# *** MUDAN√áA CR√çTICA: Converter para Int64 com suporte a nulos (pd.NA) ***
# Isso garante que 1.0 se torna 1, e os valores NaN permanecem como pd.NA
# Se converter Artigo e Numero_n no df_secundario_fund para Int64, pode
# simplificar o merge, mas aqui for√ßamos a string vazia para o merge funcionar
# com o df_secundario_fund como est√°.

df_original['Artigo'] = df_original['Artigo'].astype(pd.Int64Dtype())
df_original['Numero_n'] = df_original['Numero_n'].astype(pd.Int64Dtype())


# * Convers√£o final para string e preenchimento de nulos para o merge
# * ser consistente com o pr√©-processamento do df_secundario_fund (que √© string/vazio)
df_original['Artigo'] = df_original['Artigo'].fillna(pd.NA).astype(str).str.replace('<NA>', '').str.strip()
df_original['Numero_n'] = df_original['Numero_n'].fillna(pd.NA).astype(str).str.replace('<NA>', '').str.strip()
df_original['Alinea'] = df_original['Alinea'].fillna('').str.strip()

print("Extra√ß√£o conclu√≠da. Realizando Merge...")

# 3. Merge e Renomea√ß√£o
# Executa o mapeamento (merge) usando as 3 colunas extra√≠das como chave
df_original = pd.merge(
    left=df_original,
    right=df_secundario_fund[colunas_chave_fund + [id_coluna_secundaria]], 
    on=colunas_chave_fund,
    how='left',
)

# Renomear a coluna de ID resultante do merge
df_original.rename(columns={id_coluna_secundaria: 'id_fundamentacao'}, inplace=True)


# 4. Remo√ß√£o das colunas tempor√°rias E DA COLUNA ORIGINAL
print(f"üëâ 2/4. Removendo as colunas tempor√°rias {novas_colunas_temp} e a coluna original '{coluna_original}'...")
colunas_a_remover = novas_colunas_temp + [coluna_original]
df_original = df_original.drop(columns=colunas_a_remover)


print("-" * 50)
print("‚úÖ Transforma√ß√£o de Fundamenta√ß√£o Conclu√≠da e Colunas Antigas Removidas.")
print(f"Linhas em df_original: {len(df_original)}")
print(f"Colunas finais (com o novo ID): {df_original.columns.tolist()}")
print("\nPrimeiras 5 linhas do DataFrame transformado:")
print(df_original.head())
```

```python
import pandas as pd
import numpy as np
import re # Necess√°rio para a fun√ß√£o normalizar_texto_chave
import os

# --- 1. Carregamento dos DataFrames (Corrigido) ---

coluna_acordo_original = 'DescrAcordoQuadro'

# üéØ CORRE√á√ÉO: Carregar df_original (ContratosPublicos2024.xlsx)
df_original.columns = df_original.columns.str.strip()
   


# Carregar df_secundario_acordo ('tabelas2/DescrAcordoQuadro.csv')
try:
    df_secundario_acordo = pd.read_csv('tabelas2/DescrAcordoQuadro.csv') 
    id_coluna_secundaria = 'ID' 
    coluna_chave_secundaria = 'DescrAcordoQuadro'
    # üéØ Adicionar convers√£o para Int64Dtype no ID para evitar 1.0 no ID final
    df_secundario_acordo[id_coluna_secundaria] = df_secundario_acordo[id_coluna_secundaria].astype(pd.Int64Dtype())
    print("‚úÖ DataFrame de lookup carregado.")
except FileNotFoundError:
    print("‚ùå ERRO: O arquivo de lookup n√£o foi encontrado.")
    exit()

print("-" * 50)

# =========================================================================
# === CORRE√á√ÉO CR√çTICA: NORMALIZA√á√ÉO DE CHAVES PARA MAPPING ===
# =========================================================================

# Fun√ß√£o de limpeza: Remove pontua√ß√£o, acentos, converte para min√∫sculas e remove espa√ßos extras
def normalizar_texto_chave(texto):
    if pd.isna(texto):
        return texto  # Mant√©m NaN
    texto = str(texto).lower()
    # Remove pontua√ß√£o, exceto letras e n√∫meros
    texto = re.sub(r'[^a-z0-9\s]', '', texto) 
    # Remove espa√ßos duplos
    texto = re.sub(r'\s+', ' ', texto) 
    return texto.strip()

print("üëâ 1/3. Normalizando o texto das chaves (Fuzzy Matching)..")

# 1. Pr√©-processamento das chaves de mapeamento

# No df_original: Aplica a normaliza√ß√£o na coluna para criar a chave de merge
df_original['chave_acordo_normalizada'] = df_original[coluna_acordo_original].apply(normalizar_texto_chave)

# No df_secundario: Aplica a normaliza√ß√£o na coluna chave
df_secundario_acordo['chave_secundaria_normalizada'] = df_secundario_acordo[coluna_chave_secundaria].apply(normalizar_texto_chave)


# 2. Merge (Mapeamento pelo Nome Normalizado)
print("üëâ 2/3. Realizando o Merge com as chaves normalizadas...")

# Faz o merge usando as novas colunas normalizadas como chaves
df_original = pd.merge(
    left=df_original,
    right=df_secundario_acordo[[id_coluna_secundaria, 'chave_secundaria_normalizada']], 
    left_on='chave_acordo_normalizada',
    right_on='chave_secundaria_normalizada',
    how='left',
)

# Renomear a coluna de ID resultante do merge
df_original.rename(columns={id_coluna_secundaria: 'id_descrAcordoQuadro'}, inplace=True)

# üéØ Convers√£o final do ID para inteiro (sempre boa pr√°tica para IDs)
df_original['id_descrAcordoQuadro'] = df_original['id_descrAcordoQuadro'].astype(pd.Int64Dtype())


# 3. Limpeza Final e Remo√ß√£o de Colunas
print("üëâ 3/3. Removendo colunas tempor√°rias...")

# Colunas a remover: ambas as chaves normalizadas e a chave secund√°ria trazida pelo merge
# NOTA: N√£o removemos a coluna original 'DescrAcordoQuadro' neste caso, pois ela pode ser √∫til para ver o nome original do Acordo.
colunas_a_remover = ['chave_acordo_normalizada', 'chave_secundaria_normalizada', coluna_chave_secundaria]
df_original = df_original.drop(columns=colunas_a_remover, errors='ignore')


print("-" * 50)
print("‚úÖ Transforma√ß√£o da Coluna DescrAcordoQuadro Conclu√≠da.")
print(f"Linhas em df_original: {len(df_original)}")
print(f"Colunas finais (com o novo ID): {df_original.columns.tolist()}")
print("\nPrimeiras 5 linhas do DataFrame transformado:")
print(df_original.head(5))
```

```python
import pandas as pd
import numpy as np
import re

# --- 1. Configura√ß√£o e Assun√ß√µes ---

# Defini√ß√£o CORRETA dos nomes das colunas baseada na imagem anterior
colunas_entidade_original = ['adjudicante', 'adjudicatarios'] 

# Supondo que df_original est√° carregado e padronizado (min√∫sculas)
try:
    # Garantimos que a padroniza√ß√£o para min√∫sculas est√° feita para evitar KeyError
    df_original.columns = df_original.columns.str.lower()
    df_original.columns = df_original.columns.str.strip()
    print("‚úÖ Nomes das colunas do df_original padronizados.")
except NameError:
    print("‚ùå ERRO: O DataFrame 'df_original' n√£o est√° carregado. Por favor, carregue-o.")
    exit()

print("-" * 50)

# =========================================================================
# === FUN√á√ÉO DE EXTRA√á√ÉO DO NIF (Reutilizada) ===
# =========================================================================

def extrair_nif(texto):
    """Extrai a sequ√™ncia num√©rica (NIF) antes do primeiro h√≠fen ou espa√ßo."""
    if pd.isna(texto):
        return np.nan
    texto_str = str(texto).strip()
    
    # Tenta encontrar a sequ√™ncia num√©rica no in√≠cio seguida de espa√ßo/h√≠fen
    match = re.match(r'(\d+)\s*-', texto_str)
    if match:
        return match.group(1).strip()
    
    # Tenta pegar tudo antes do primeiro h√≠fen
    partes = texto_str.split('-', 1)
    if partes and partes[0].isdigit():
        return partes[0].strip()
        
    return np.nan

# =========================================================================
# === TRANSFORMA√á√ÉO: SUBSTITUIR PELO NIF ===
# =========================================================================

for coluna in colunas_entidade_original: 
    print(f"üëâ Processando coluna: {coluna}...")
    
    # 1. Aplica a fun√ß√£o de extra√ß√£o diretamente na coluna
    df_original[coluna] = df_original[coluna].apply(extrair_nif)
    
    # 2. Converte para string (garantir que o NIF n√£o seja tratado como n√∫mero se tiver 8 d√≠gitos)
    df_original[coluna] = df_original[coluna].fillna('').astype(str).str.strip()
    
    # 3. Substitui a string 'nan' gerada pelo fillna (se a coluna original estava vazia) por vazio
    df_original[coluna] = df_original[coluna].replace('nan', '')
    
    print(f"   ‚úÖ Coluna '{coluna}' agora cont√©m apenas o NIF.")


print("-" * 50)
print("‚úÖ Transforma√ß√£o de Entidades para apenas NIF Conclu√≠da.")
print(f"Linhas em df_original: {len(df_original)}")
print("\nPrimeiras 5 linhas do DataFrame transformado:")
print(df_original[colunas_entidade_original].head())
```

```python
colunas_a_remover = 'localexecucao'
df_original = df_original.drop(columns=colunas_a_remover, errors='ignore')
nome_ficheiro_saida = 'Contratos.csv'
    # Guarda o DataFrame no formato CSV.
    # index=False garante que a coluna de √≠ndice do Pandas n√£o √© escrita no ficheiro.
df_original.to_csv(nome_ficheiro_saida, index=False)
```
